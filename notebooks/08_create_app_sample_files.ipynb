1. Title and objective

# CICIoMT2024 - Create Sample CSV Files for Streamlit App

## Objective
This notebook creates sample CSV files for the Streamlit application.

It produces:
- a header-only template file for schema reference
- example CSV files with real rows for testing
- one sample file for each app mode:
  - Binary Detection
  - Grouped Multiclass
  - Full Multiclass

2. Imports

```python
from pathlib import Path
import pandas as pd
```

3. Define paths

```python
BASE_DIR = Path("..")
APP_DIR = BASE_DIR / "app"
PROCESSED_DIR = BASE_DIR / "data" / "processed"
REPORT_DIR = BASE_DIR / "reports" / "results"

APP_DIR.mkdir(parents=True, exist_ok=True)

print("App dir:", APP_DIR.resolve())
print("Processed dir:", PROCESSED_DIR.resolve())
print("Report dir:", REPORT_DIR.resolve())
```

4. Load saved feature lists

```python
binary_features = pd.read_csv(REPORT_DIR / "binary_rf_features.csv")["feature"].tolist()
grouped_features = pd.read_csv(REPORT_DIR / "grouped_rf_features.csv")["feature"].tolist()
full_features = pd.read_csv(REPORT_DIR / "full_rf_features.csv")["feature"].tolist()

print("Binary feature count:", len(binary_features))
print("Grouped feature count:", len(grouped_features))
print("Full feature count:", len(full_features))
```

5. Load processed datasets

```python
X_test_binary = pd.read_csv(PROCESSED_DIR / "X_test_tree_binary.csv")
X_test_grouped = pd.read_csv(PROCESSED_DIR / "X_test_tree_grouped.csv")
X_test_full = pd.read_csv(PROCESSED_DIR / "X_test_tree_full.csv")

print("X_test_binary:", X_test_binary.shape)
print("X_test_grouped:", X_test_grouped.shape)
print("X_test_full:", X_test_full.shape)
```

6. Create header-only templates

## Header-only templates
These files contain only column names and no data rows.
They are useful for showing the required schema.

```python
binary_template = pd.DataFrame(columns=binary_features)
grouped_template = pd.DataFrame(columns=grouped_features)
full_template = pd.DataFrame(columns=full_features)

binary_template.to_csv(APP_DIR / "binary_input_template.csv", index=False)
grouped_template.to_csv(APP_DIR / "grouped_input_template.csv", index=False)
full_template.to_csv(APP_DIR / "full_input_template.csv", index=False)

print("Header-only templates saved.")
```

7. Create real example sample files

## Example sample files with real rows
These files contain actual rows from the processed test datasets and can be uploaded directly into the Streamlit app.

```python
n_rows = 5

binary_example = X_test_binary[binary_features].head(n_rows).copy()
grouped_example = X_test_grouped[grouped_features].head(n_rows).copy()
full_example = X_test_full[full_features].head(n_rows).copy()

binary_example.to_csv(APP_DIR / "binary_input_example.csv", index=False)
grouped_example.to_csv(APP_DIR / "grouped_input_example.csv", index=False)
full_example.to_csv(APP_DIR / "full_input_example.csv", index=False)

print("Example files with real rows saved.")
```

8. Create one generic default sample file

## Generic default sample
This is a convenient default sample file for the app, based on the grouped multiclass feature set.

```python
grouped_example.to_csv(APP_DIR / "sample_input_template.csv", index=False)
print("Saved app/sample_input_template.csv with real rows.")
```

## Optional labeled demo files
These files include the true labels for reference, but they should not be uploaded to the app directly unless you remove the label columns first.

```python
y_test_binary = pd.read_csv(PROCESSED_DIR / "y_test_binary.csv").squeeze("columns")
y_test_grouped = pd.read_csv(PROCESSED_DIR / "y_test_grouped.csv").squeeze("columns")
y_test_full = pd.read_csv(PROCESSED_DIR / "y_test_full.csv").squeeze("columns")
```

```python
binary_demo_labeled = X_test_binary[binary_features].head(n_rows).copy()
binary_demo_labeled["true_label"] = y_test_binary.head(n_rows).values

grouped_demo_labeled = X_test_grouped[grouped_features].head(n_rows).copy()
grouped_demo_labeled["true_label"] = y_test_grouped.head(n_rows).values

full_demo_labeled = X_test_full[full_features].head(n_rows).copy()
full_demo_labeled["true_label"] = y_test_full.head(n_rows).values

binary_demo_labeled.to_csv(APP_DIR / "binary_input_example_labeled.csv", index=False)
grouped_demo_labeled.to_csv(APP_DIR / "grouped_input_example_labeled.csv", index=False)
full_demo_labeled.to_csv(APP_DIR / "full_input_example_labeled.csv", index=False)

print("Optional labeled demo files saved.")
```

10. Show preview of created files

```python
print("Binary example preview:")
display(binary_example.head())

print("Grouped example preview:")
display(grouped_example.head())

print("Full example preview:")
display(full_example.head())
```
